***Step 2: Data Pre-processing***

Data pre-processing is an essential step in any data analysis workflow. It ensures the dataset is clean, consistent, and ready for deeper analysis. In this section, we perform several important operations to improve the quality of our data.

**Cleaning Column Values & Correcting Data Types**

Real-world datasets often contain inconsistent, messy, or incorrectly formatted values. These issues may arise from manual data entry, file formatting differences, or data exported from external sources. Before performing any analysis, it is essential to clean column values and correct data types to ensure the dataset is accurate and usable.

Cleaning column values helps remove:

*   Extra spaces, tabs, and newline characters
*   Hidden non-breaking spaces (\xa0)
*   Unwanted characters and symbols
*   Text mixed with numeric values

This step makes the data consistent, easier to read, and ready for proper numeric conversion and analysis.

Datasets often contain incorrect or inconsistent data types—for example:

*  Dates stored as strings
*  Numbers stored as text
*  Categorical fields mistakenly treated as numerical

Correcting data types allows efficient filtering, grouping, and computations.

One of the colunms is totally blank in the data set. Removed the unwanted data from the data set.

```python
# df.drop(df.columns[17], axis=1, inplace=True)

df.drop(columns=['stock_days'], axis=1, inplace=True)

print("The shape (rows, columns) of the dataframe after removing missing datas:")
df.shape
```

```python
df.columns = df.columns.str.replace('_', ' ').str.title()

print("Step 1 completed: Column names cleaned!\n")
print(df.columns.tolist())
```

```python
numeric_cols = [
    "State Code", "Capacity", "Daily Requirement", "Daily Receipt",
    "Daily Consumption", "Req Normative Stock", "Normative Stock Days", "Indigenous Stock",
    "Import Stock", "Total Stock", "Plf Prcnt", "Actual Vs Normative Stock Prcnt"
]
for col in numeric_cols:
    df[col] = ( df[col].astype(str)
                .str.replace(r'[^0-9.-]', '', regex=True)
                .replace('', np.nan)
                .astype(float)  )

print("\nStep 2 completed: Numeric columns cleaned and converted to float!\n")
print(df[numeric_cols].dtypes)

```

```python
text_cols = df.columns.difference(numeric_cols)

for col in text_cols:
    df[col] = (
        df[col].astype(str)
                .str.replace('\xa0', ' ', regex=False)
                .str.replace(r'\s+', ' ', regex=True)
                .str.strip()
    )

print("\nStep 3 completed: Text columns cleaned and formatted!\n")
print(df[text_cols].head())
```

```python
df = df.convert_dtypes()

print("\nStep 4 completed: Data types corrected!\n")
print(df.dtypes)
```

```python

print("\nFinal Cleaned Dataset Preview:")
df.head()
```

**Derived column using date**

To improve temporal analysis and identify patterns over time, we create a derived column called “Month” from the existing Date column. This derived column extracts the month value from each date, allowing records to be grouped by month. By doing this, we can more easily analyze monthly trends, compare stock levels and consumption across different periods, and visualize seasonal variations in coal availability and usage.

The transformation is performed using pandas’ datetime functions, ensuring accurate and consistent date-based analysis across the dataset.

```python
# Convert to datetime (keep this)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Month'] = df['Date'].dt.month_name()

# FORMAT Date for display (DD-MM-YYYY)
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')

print(df[['Date', 'Month']].head())
print(df['Date'].dtype)
```

**Bivariate Analysis**

Bivariate analysis studies the relationship between two variables at a time. It reveals how one feature influences or interacts with another.

7) **State-wise Daily Consumption vs Daily Requirement**

The distribution between daily consumption and requirement is nearly balanced, showing efficient alignment between demand planning and actual usage. This indicates better operational control and optimized coal utilization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("daily-coal-stocks filtered.csv")
```

```python
state_summary = df.groupby("State Name")[["Daily Consumption", "Daily Requirement"]].sum().reset_index()
for _, row in state_summary.iterrows():
  plt.figure(figsize=(4,4))
  plt.pie(
        [row["Daily Consumption"], row["Daily Consumption"]],
        labels=["Daily Consumption", "Daily Requirement"],
        autopct='%1.1f%%',
        startangle=90)
  plt.title(f"{row['State Name']} : Daily Consumption vs Daily Requirement", fontweight='bold')
  plt.show()
```

10) **Average Total Stock by State**

11) **Stock Comparison**

The pairplot visualizes pairwise relationships between normative stock requirements,indigenous and import coal stocks. It helps identify correlations, trends, and distribution patterns across these variables in a single view.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("daily-coal-stocks filtered.csv")

# Select required columns
cols = [
    "Req Normative Stock",
    "Indigenous Stock",
    "Import Stock"
]

# Create pairplot
sns.pairplot(df[cols], diag_kind="kde")
plt.show()
```

12) **Correlation Heatmap of Numeric Variables**

The heatmap shows that variables like Capacity, Daily Requirement, Daily Consumption, and Req Normative Stock are highly positively correlated (values close to 0.9), indicating these metrics tend to increase together.

Variables like Normative Stock Days have negative correlations with many other numeric variables (blue shades), meaning as stock days increase, some other quantities like Daily Consumption tend to decrease.



```python
plt.figure(figsize=(14,10))

corr = df.select_dtypes(include='number').corr()

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    square=True,
    annot_kws={"size": 9}
)

plt.title('Correlation Heatmap of Numeric Variables', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
